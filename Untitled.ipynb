```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data=pd.read_csv('dummy_sensor_data.csv')
data.head(10)
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'], format='%d-%m-%Y %H:%M')
```

```python
data['Date'] = data['Timestamp'].dt.date
data['Time'] = data['Timestamp'].dt.time
```

```python
data = data.drop(columns=['Timestamp'])
```

```python
data.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def calculate_mae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Assuming your data is stored in a DataFrame named 'data'
X = data.drop(['Reading'], axis=1)

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X)

y = data['Reading']

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Define the Random Forest Regressor model
rf_model = RandomForestRegressor(random_state=42)

# Define hyperparameters for tuning
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [25],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, 
                           scoring='neg_mean_absolute_error', cv=3)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Train the model with the best hyperparameters
best_rf_model = RandomForestRegressor(random_state=42, **best_params)
best_rf_model.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = best_rf_model.predict(X_test)

# Evaluate the model on the validation set
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error on Validation Set:", mae)

# Calculate and print Mean Absolute Percentage Error (MAPE)
mape = calculate_mape(y_test, y_pred)
print(f"Random Forest Model MAPE: {mape:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Assuming your data is stored in a DataFrame named 'data'
X = data.drop(['Reading'], axis=1)

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X)

y = data['Reading']

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Define the XGBoost Regressor model
xgb_model = XGBRegressor(random_state=42)

# Define hyperparameters for tuning
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
}

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid,
                           scoring='neg_mean_absolute_error', cv=3)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Train the model with the best hyperparameters
best_xgb_model = XGBRegressor(random_state=42, **best_params)
best_xgb_model.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = best_xgb_model.predict(X_test)

# Evaluate the model on the validation set
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error on Validation Set:", mae)

# Calculate and print Mean Absolute Percentage Error (MAPE)
mape = calculate_mape(y_test, y_pred)
print(f"XGBoost Model MAPE: {mape:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_absolute_error

def calculate_mae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

# Extract relevant columns
time_series_data = data[['Date', 'Time', 'Reading']]

# Combine Date and Time columns into a single datetime column
time_series_data['Datetime'] = pd.to_datetime(time_series_data['Date'].astype(str) + ' ' + time_series_data['Time'].astype(str))
time_series_data = time_series_data.drop(['Date', 'Time'], axis=1)

# Sort the data by datetime
time_series_data = time_series_data.sort_values('Datetime')

# Set Datetime as the index
time_series_data = time_series_data.set_index('Datetime')

# Resample the data to a regular time frequency (e.g., hourly)
time_series_data_resampled = time_series_data.resample('H').mean().interpolate()

# Normalize the data using Min-Max scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(time_series_data_resampled)

# Split the data into training and testing sets
train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[0:train_size], scaled_data[train_size:]

print("Shape of training data : ",train_data.shape)

# Function to create sequences for the LSTM model
def create_sequences(data, sequence_length):
    sequences = []
    targets = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i+sequence_length]
        target = data[i+sequence_length]
        sequences.append(seq)
        targets.append(target)
    return np.array(sequences), np.array(targets)

# Define sequence length (adjust based on your data characteristics)
sequence_length = 24  # For example, using the past 24 hours to predict the next hour

# Create sequences for training
X_train, y_train = create_sequences(train_data, sequence_length)

# Create sequences for testing
X_test, y_test = create_sequences(test_data, sequence_length)

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Evaluate the model on the test set
y_pred = model.predict(X_test)
y_pred_inverse = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test_inverse = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate and print Mean Absolute Error (MAE) or other relevant metrics
mae = calculate_mae(y_test_inverse, y_pred_inverse)
print("Mean Absolute Error on Test Set:", mae)

import joblib

joblib.dump(model, "app/model.pkl")
```
